```python
merged_df = RA_BlackorAfricanAmerican_df_copy.merge(RA_HispanicORlatino_df_copy, on='Geographic Area Name').merge(RA_White_df_copy, on='Geographic Area Name').merge(RA_Asian_df_copy, on='Geographic Area Name')
merged_df
```

```python
def calculate_proportions(row):
    row_sum = row.sum()
    if row_sum != 0:
        return row / row_sum
    else:
        return row

# Apply the function to each row
proportion_df = merged_df.apply(calculate_proportions, axis=1)

# Display the resulting DataFrame
proportion_df
```

```python
proportion_df=proportion_df.reset_index()

```

```python
proportion_df['County'] = proportion_df['Geographic Area Name'].str.split(',').str[0]
proportion_df['County'] = proportion_df['County'].str.replace(' County', '')
proportion_df
```

```python
RA_Georgia=pd.read_csv('../../Obstetric accessibility visualization/Georgia/BG_County_numbers.csv')
```

```python
merged_df = pd.merge(RA_Georgia, proportion_df, on='County')
merged_df
```

```python
columns_to_multiply = [
    'Total BlackorAfricanAmerican Female, 18-44', 
    'Total HispanicORlatino Female, 18-44', 
    'Total White Female, 18-44', 
    'Total Asian Female, 18-44'
]

# Performing the multiplication
for column in columns_to_multiply:
    merged_df[column] = (merged_df[column] * merged_df['Population'])

# Display the modified DataFrame
merged_df
```

<h1>Reproductive Aged women across various races at Block group level</h1>

```python
columns_to_round = [
    'Total BlackorAfricanAmerican Female, 18-44', 
    'Total HispanicORlatino Female, 18-44', 
    'Total White Female, 18-44', 
    'Total Asian Female, 18-44'
]

for index, row in merged_df.iterrows():
    # Round all but the last value
    for column in columns_to_round[:-1]:
        merged_df.at[index, column] = round(row[column])
    
    # Calculate the last value to ensure the sum matches the Population
    last_column = columns_to_round[-1]
    merged_df.at[index, last_column] = row['Population'] - sum([round(row[col]) for col in columns_to_round[:-1]])

merged_df
```
