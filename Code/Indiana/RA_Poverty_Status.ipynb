<h1>Poverty Status on a block group Level using Census Tract Data ¶</h1>

```python
def calculate_proportions(row):
    total = row['Income in the past 12 months below poverty level!!Female!!18 to 44 years'] + row['Income in the past 12 months at or above poverty level!!Female!!18 to 44 years']
    if total != 0:
        below_poverty = row['Income in the past 12 months below poverty level!!Female!!18 to 44 years'] / total
        above_poverty = row['Income in the past 12 months at or above poverty level!!Female!!18 to 44 years'] / total
    else:
        below_poverty = 0
        above_poverty = 0
    return pd.Series([below_poverty, above_poverty], index=['Proportion Below Poverty', 'Proportion Above Poverty'])

# Apply the function to each row
proportion_df = PovertyStatusbyCensusTract_df.apply(calculate_proportions, axis=1)

# Combine the new proportions with the original DataFrame
result_df = pd.concat([PovertyStatusbyCensusTract_df, proportion_df], axis=1)

# Display the resulting DataFrame
result_df
result_df.drop(['Income in the past 12 months below poverty level!!Female!!18 to 44 years', 
         'Income in the past 12 months at or above poverty level!!Female!!18 to 44 years'], 
        axis=1, inplace=True)

# Display the updated DataFrame
result_df
```

```python
RA_Iowa_BG = pd.read_csv('../../Obstetric accessibility visualization/Indiana/RA_Count_BlockGroup.csv')
#RA_Iowa_BG.rename(columns=RA_Iowa_BG.iloc[0], inplace=True)
RA_Iowa_BG_df = RA_Iowa_BG[[
    ''
    'Estimate!!Total:!!Female:!!18 and 19 years',
    'Estimate!!Total:!!Female:!!20 years',
    'Estimate!!Total:!!Female:!!21 years',
    'Estimate!!Total:!!Female:!!22 to 24 years',
    'Estimate!!Total:!!Female:!!25 to 29 years',
    'Estimate!!Total:!!Female:!!30 to 34 years', 
    'Estimate!!Total:!!Female:!!35 to 39 years',
    'Estimate!!Total:!!Female:!!40 to 44 years'
]]

RA_Iowa_BG_df_copy = RA_Iowa_BG_df.copy()
RA_Iowa_BG_df_copy.drop(RA_Iowa_BG_df.index[0], inplace=True)
age_columns = [
    'Estimate!!Total:!!Female:!!18 and 19 years',
    'Estimate!!Total:!!Female:!!20 years',
    'Estimate!!Total:!!Female:!!22 to 24 years',
    'Estimate!!Total:!!Female:!!25 to 29 years',
    'Estimate!!Total:!!Female:!!30 to 34 years',
    'Estimate!!Total:!!Female:!!30 to 34 years',
    'Estimate!!Total:!!Female:!!35 to 39 years',
    'Estimate!!Total:!!Female:!!40 to 44 years'
]
for col in age_columns:
    if col in RA_Iowa_BG_df_copy.columns:
        RA_Iowa_BG_df_copy[col] = RA_Iowa_BG_df_copy[col].astype(str)  # Ensure it's a string
        RA_Iowa_BG_df_copy[col] = RA_Iowa_BG_df_copy[col].apply(lambda x: pd.to_numeric(x, errors='coerce'))

# Create the new column by summing the specified columns
RA_Iowa_BG_df_copy['Total Female Age 18-44 years'] = RA_Iowa_BG_df_copy[age_columns].sum(axis=1)
RA_Iowa_BG_df_copy.head()
```

```python





# Drop all other columns except 'Geographic Area Name' and the new column
RA_Iowa_BG_df_copy = RA_Iowa_BG_df_copy[['Geographic Area Name', 'Total Female Age 18-44 years']]
RA_Iowa_BG_df_copy.rename(columns={'Geographic Area Name': 'Geographic Area Names'}, inplace=True)

# Extracting the area name from the 'Geographic Area Names' column
RA_Iowa_BG_df_copy['Geographic Area Name'] = RA_Iowa_BG_df_copy['Geographic Area Names'].str.extract(r'; (.*)')

# Display the first few rows of the DataFrame
RA_Iowa_BG_df_copy.head()
merged_df = pd.merge(RA_Iowa_BG_df_copy, result_df, on='Geographic Area Name')
columns_to_multiply = [
    'Proportion Below Poverty',
       'Proportion Above Poverty'
]

# Performing the multiplication
for column in columns_to_multiply:
    merged_df[column] = (merged_df[column] * merged_df['Total Female Age 18-44 years'])
columns_to_round = ['Proportion Below Poverty',
       'Proportion Above Poverty'
    
]

for index, row in merged_df.iterrows():
    # Round all but the last value
    for column in columns_to_round[:-1]:
        merged_df.at[index, column] = round(row[column])
    
    # Calculate the last value to ensure the sum matches the Population
    last_column = columns_to_round[-1]
    merged_df.at[index, last_column] = row['Total Female Age 18-44 years'] - sum([round(row[col]) for col in columns_to_round[:-1]])
    
merged_df.drop('Geographic Area Name', axis=1, inplace=True)
merged_df.head()
#PovertyStatusbyCounty_df.to_csv('PovertyStatusbyBlockGroup_for_indiana.csv', index=False)
```
