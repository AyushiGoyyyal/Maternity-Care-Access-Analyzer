```python
import math

columns_to_round = [
    'Income in the past 12 months below poverty level!!Female!!18 to 44 years',
    'Income in the past 12 months at or above poverty level!!Female!!18 to 44 years'
]

for index, row in merged_df.iterrows():
    rounded_sum = 0

    # Round all but the last value
    for column in columns_to_round[:-1]:
        value = row[column]
        if not math.isnan(value):  # Check if the value is not NaN
            rounded_value = round(value)
            merged_df.at[index, column] = rounded_value
            rounded_sum += rounded_value
        else:
            merged_df.at[index, column] = value  # Keep the NaN or handle it differently if needed

    # Calculate the last value to ensure the sum matches the Population
    last_column = columns_to_round[-1]
    last_value = row[last_column]
    if not math.isnan(last_value):
        # Subtract the rounded sum from the total and assign it to the last column
        merged_df.at[index, last_column] = row['Total Female Age 18-44 years'] - rounded_sum
    else:
        merged_df.at[index, last_column] = last_value  # Keep the NaN or handle it differently if needed

# Drop columns as needed
merged_df.drop('Geographic Area Name', axis=1, inplace=True)
merged_df.drop('Total Female Age 18-44 years', axis=1, inplace=True)
merged_df=merged_df.drop('Geography',axis=1)
merged_df.head()
```

```python
def calculate_proportions(row):
    with_insurance = row['Total Female, 19-44 with Health Insurance']
    without_insurance = row['Total Female, 19-44 No Health Insurance']
    total = with_insurance + without_insurance

    if total != 0:
        row['Total Female, 19-44 with Health Insurance'] = with_insurance / total
        row['Total Female, 19-44 No Health Insurance'] = without_insurance / total

    return row

# Apply the function to each row
proportion_df = merged_df.apply(calculate_proportions, axis=1)

proportion_df=proportion_df.reset_index()
proportion_df

```

```python
merged_df = pd.merge(RA_Iowa_BG_df_copy, proportion_df, on='Geographic Area Name')
merged_df
```

```python
columns_to_round = [
'Total Female, 19-44 with Health Insurance',
       'Total Female, 19-44 No Health Insurance',
]

for index, row in merged_df.iterrows():
    # Round all but the last value
    for column in columns_to_round[:-1]:
        merged_df.at[index, column] = round(row[column])
    
    # Calculate the last value to ensure the sum matches the Population
    last_column = columns_to_round[-1]
    merged_df.at[index, last_column] = row['Total Female Age 18-44 years'] - sum([round(row[col]) for col in columns_to_round[:-1]])
    
merged_df.drop('Geographic Area Name', axis=1, inplace=True)
merged_df.drop('Total Female Age 18-44 years', axis=1, inplace=True)
merged_df.head()
```
