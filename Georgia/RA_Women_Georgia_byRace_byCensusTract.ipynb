```python
import pandas as pd
RA_Georgia_BG=pd.read_csv('../../Obstetric accessibility visualization/Indiana/RA_NativeHawaiin_censusTract.csv')
RA_Georgia_BG.head()
RA_Georgia_BG.rename(columns=RA_Georgia_BG.iloc[0], inplace=True)
RA_Georgia_BG.columns
RA_Georgia_BG_df = RA_Georgia_BG[[
    'Geographic Area Name',
    'Estimate!!Total!!Female!!18 and 19 years',
    'Estimate!!Total!!Female!!20 years',
    'Estimate!!Total!!Female!!22 to 24 years',
    'Estimate!!Total!!Female!!25 to 29 years',
    'Estimate!!Total!!Female!!30 to 34 years',
    'Estimate!!Total!!Female!!30 to 34 years',
    'Estimate!!Total!!Female!!35 to 39 years',
    'Estimate!!Total!!Female!!40 to 44 years'
]]

RA_Georgia_BG_df_copy = RA_Georgia_BG_df.copy()
RA_Georgia_BG_df_copy.drop(RA_Georgia_BG_df.index[0], inplace=True)
age_columns = [
    'Estimate!!Total!!Female!!18 and 19 years',
    'Estimate!!Total!!Female!!20 years',
    'Estimate!!Total!!Female!!22 to 24 years',
    'Estimate!!Total!!Female!!25 to 29 years',
    'Estimate!!Total!!Female!!30 to 34 years',
    'Estimate!!Total!!Female!!30 to 34 years',
    'Estimate!!Total!!Female!!35 to 39 years',
    'Estimate!!Total!!Female!!40 to 44 years'
]

RA_Georgia_BG_df_copy
```

```python
for col in age_columns:
    if col in RA_Georgia_BG_df_copy.columns:
        RA_Georgia_BG_df_copy[col] = RA_Georgia_BG_df_copy[col].astype(str)  # Ensure it's a string
        RA_Georgia_BG_df_copy[col] = RA_Georgia_BG_df_copy[col].apply(lambda x: pd.to_numeric(x, errors='coerce'))

# Create the new column by summing the specified columns
RA_Georgia_BG_df_copy['Total Female Age 18-44 years'] = RA_Georgia_BG_df_copy[age_columns].sum(axis=1)

# Drop all other columns except 'Geographic Area Name' and the new column
RA_Georgia_BG_df_copy = RA_Georgia_BG_df_copy[['Geographic Area Name', 'Total Female Age 18-44 years']]
RA_Georgia_BG_df_copy.rename(columns={'Geographic Area Name': 'Geographic Area Names'}, inplace=True)
# Display the resulting DataFrame
RA_Georgia_BG_df_copy

RA_Georgia_BG_df_copy['Geographic Area Name'] = RA_Georgia_BG_df_copy['Geographic Area Names'].str.extract(r'Block Group \d+, (.*)')

RA_Georgia_BG_df_copy.head()
```

```python
merged_df = pd.merge(RA_Georgia_BG_df_copy, proportion_df, on='Geographic Area Name')
merged_df.head()
```

```python
columns_to_multiply = [
    'Total AmericanIndianORAlaskaNative Female, 18-44',
       'Total BlackorAfricanAmerican Female, 18-44',
       'Total HispanicORlatino Female, 18-44', 'Total White Female, 18-44',
       'Total Asian Female, 18-44',
       'Total NativeHawaiin&OtherPacificIslander Female, 18-44'
]

# Performing the multiplication
for column in columns_to_multiply:
    merged_df[column] = (merged_df[column] * merged_df['Total Female Age 18-44 years'])

# Display the modified DataFrame
merged_df
```

```python
columns_to_round = [
    'Total AmericanIndianORAlaskaNative Female, 18-44',
       'Total BlackorAfricanAmerican Female, 18-44',
       'Total HispanicORlatino Female, 18-44', 'Total White Female, 18-44',
       'Total Asian Female, 18-44',
       'Total NativeHawaiin&OtherPacificIslander Female, 18-44'
]

for index, row in merged_df.iterrows():
    # Round all but the last value
    for column in columns_to_round[:-1]:
        merged_df.at[index, column] = round(row[column])
    
    # Calculate the last value to ensure the sum matches the Population
    last_column = columns_to_round[-1]
    merged_df.at[index, last_column] = row['Total Female Age 18-44 years'] - sum([round(row[col]) for col in columns_to_round[:-1]])
    
merged_df.drop('Geographic Area Name', axis=1, inplace=True)
```
