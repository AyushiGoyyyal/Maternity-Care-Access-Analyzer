<h1> RA Women in Indiana by Block Group</h1>

```python
def calculate_proportions(row):
    row_sum = row.sum()
    if row_sum != 0:
        return row / row_sum
    else:
        return row

# Apply the function to each row
proportion_df = merged_df.apply(calculate_proportions, axis=1)

# Display the resulting DataFrame
proportion_df
```

```python
proportion_df=proportion_df.reset_index()


proportion_df
```

```python
RA_Indiana_BG = pd.read_csv('../../Obstetric accessibility visualization/Indiana/RA_Count_BlockGroup.csv')

#RA_Indiana_BG.rename(columns=RA_Indiana_BG.iloc[0], inplace=True)
RA_Indiana_BG.head()
```

```python


# Assuming you have a similar file structure for Indiana
RA_Indiana_BG = pd.read_csv('../../Obstetric accessibility visualization/Indiana/RA_Count_BlockGroup.csv')
RA_Indiana_BG.head()
#RA_Indiana_BG.rename(columns=RA_Indiana_BG.iloc[0], inplace=True)
RA_Indiana_BG.columns

RA_Indiana_BG_df = RA_Indiana_BG[[
    'Geographic Area Name',
    'Estimate!!Total:!!Female:!!18 and 19 years',
    'Estimate!!Total:!!Female:!!20 years',
    'Estimate!!Total:!!Female:!!21 years',
    'Estimate!!Total:!!Female:!!22 to 24 years',
    'Estimate!!Total:!!Female:!!25 to 29 years',
    'Estimate!!Total:!!Female:!!30 to 34 years',  
    'Estimate!!Total:!!Female:!!35 to 39 years',
    'Estimate!!Total:!!Female:!!40 to 44 years'
]]

RA_Indiana_BG_df_copy = RA_Indiana_BG_df.copy()
RA_Indiana_BG_df_copy.drop(RA_Indiana_BG_df.index[0], inplace=True)

age_columns = [
    'Estimate!!Total:!!Female:!!18 and 19 years',
    'Estimate!!Total:!!Female:!!20 years',
    'Estimate!!Total:!!Female:!!21 years',
    'Estimate!!Total:!!Female:!!22 to 24 years',
    'Estimate!!Total:!!Female:!!25 to 29 years',
    'Estimate!!Total:!!Female:!!30 to 34 years', 
    'Estimate!!Total:!!Female:!!35 to 39 years',
    'Estimate!!Total:!!Female:!!40 to 44 years'
]

# Your operations on RA_Indiana_BG_df_copy here
RA_Indiana_BG_df_copy
```

```python
for col in age_columns:
    if col in RA_Indiana_BG_df_copy.columns:
        RA_Indiana_BG_df_copy[col] = RA_Indiana_BG_df_copy[col].astype(str)  # Ensure it's a string
        RA_Indiana_BG_df_copy[col] = RA_Indiana_BG_df_copy[col].apply(lambda x: pd.to_numeric(x, errors='coerce'))

# Create the new column by summing the specified columns
RA_Indiana_BG_df_copy['Total Female Age 18-44 years'] = RA_Indiana_BG_df_copy[age_columns].sum(axis=1)

# Drop all other columns except 'Geographic Area Name' and the new column
RA_Indiana_BG_df_copy = RA_Indiana_BG_df_copy[['Geographic Area Name', 'Total Female Age 18-44 years']]
RA_Indiana_BG_df_copy.rename(columns={'Geographic Area Name': 'Geographic Area Names'}, inplace=True)

# Extracting the area name from the 'Geographic Area Names' column
RA_Indiana_BG_df_copy['Geographic Area Name'] = RA_Indiana_BG_df_copy['Geographic Area Names'].str.extract(r'; (.*)')

# Display the first few rows of the DataFrame
RA_Indiana_BG_df_copy.head()
```

```python
merged_df = pd.merge(RA_Indiana_BG_df_copy, proportion_df, on='Geographic Area Name')
columns_to_multiply = [
    'Total Native Hawaiian & Other Pacific Islander Female, 18-44',
       'Total Asian Female, 18-44',
       'Total American Indian and Alaska Native Female, 18-44', 
       'Total White Female, 18-44',
'Total Black or African American Female, 18-44'
]

# Performing the multiplication
for column in columns_to_multiply:
    merged_df[column] = (merged_df[column] * merged_df['Total Female Age 18-44 years'])

```

```python
columns_to_round = [
    'Total Native Hawaiian & Other Pacific Islander Female, 18-44',
       'Total Asian Female, 18-44',
       'Total American Indian and Alaska Native Female, 18-44', 
       'Total White Female, 18-44',
    'Total Black or African American Female, 18-44'
]

for index, row in merged_df.iterrows():
    # Round all but the last value
    for column in columns_to_round[:-1]:
        merged_df.at[index, column] = round(row[column])
    
    # Calculate the last value to ensure the sum matches the Population
    last_column = columns_to_round[-1]
    merged_df.at[index, last_column] = row['Total Female Age 18-44 years'] - sum([round(row[col]) for col in columns_to_round[:-1]])
    
merged_df.drop('Geographic Area Name', axis=1, inplace=True)
merged_df
```
